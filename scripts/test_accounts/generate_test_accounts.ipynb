```python
import json
import secrets
from nacl.signing import SigningKey

NUM_ACCOUNTS = 8
BALANCE_MIN = 0
BALANCE_MAX = 1000

accounts = []
for i in range(NUM_ACCOUNTS):
    # Generate a 32-byte seed and corresponding ed25519 keypair
    seed = secrets.token_hex(32)
    sk = SigningKey(bytes.fromhex(seed))
    vk = sk.verify_key

    account = {
        "priv_key": seed,                              # 64 hex chars
        "pub_key": vk.encode().hex(),                  # 64 hex chars
        "balance": secrets.randbelow(BALANCE_MAX - BALANCE_MIN + 1) + BALANCE_MIN,
        "nonce": 0
    }
    accounts.append(account)

# Write to JSON file
with open("accounts.json", "w") as f:
    json.dump(accounts, f, indent=2)

# Also print to stdout
print(json.dumps(accounts, indent=2))
```
